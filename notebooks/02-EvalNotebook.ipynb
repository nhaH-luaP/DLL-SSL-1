```python
import json
import matplotlib.pyplot as plt
```

```python
with open('./example_outputs/first_test_results.json', 'r') as file:
    results = json.load(file)
```

```python
train_metric = 'val_loss'

values = []
for epoch_results in results['train_metrics']:
    values.append(epoch_results[train_metric])
plt.plot(range(len(values)), values)
plt.xlabel('epoch')
plt.ylabel(train_metric)
plt.show()
```
